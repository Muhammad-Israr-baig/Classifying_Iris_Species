```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

# Step 1: Load the dataset
iris = load_iris()

# Step 2: Split data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(
    iris.data, iris.target, random_state=0
)

# Step 3: Create the model
knn = KNeighborsClassifier(n_neighbors=1)

# Step 4: Train (fit) the model
knn.fit(X_train, y_train)

# Step 5: Test accuracy
print("Accuracy on test set:", knn.score(X_test, y_test))
```

```python
import numpy as np
X_new = np.array([[5,2.9,1,2]])
print("X_new.shape: {}" .format(X_new.shape))
```

```python
prediction = knn.predict(X_new)
print("Prediction: {}".format(prediction))
print("prediction target name: {}".format(iris['target_names'][prediction[0]]))
```

```python
y_pred = knn.predict(X_test)
print("test set predictions: \n {}". format(y_pred))
```

```python
import numpy as np
result= np.mean(y_pred == y_test) 
print("test set score",result)
```
